```python

matriz =[
    [2,5,1],            #fila 0
    [0,3,7],            #fila 1
    [4,6,9]             #fila 2
    
]

filas = 0               #variable para contar cuantas filas tiene la matriz
for _ in matriz:        #recorre cada fila de matriz 
    filas +=1           #cada fila incrementa el contador de filas
    
cols = 0                #variablepara contar cuantas columnas tiene la matriz

for _ in matriz[0]:     #recorro la primera fila para contar kas columnas 
    cols+=1 
    
suma_filas = [0]*filas  #creamos un nuevo vector para gauradar la suma de cada fila

i=0                     #indice de filas 

while i<filas:          #recorremos las filas una por una 
    suma = 0                #acumulador para la suma de la fila i
    j=0                         #indice de columnas
    while j<cols:               #recprre todas las columanas de la fila i
        suma += matriz[i][j]    #sumo el elemento de la posicion (i,j)
        j+=1                    #paso a la siguente columna 
    suma_filas[i]=suma          #guardo la suma total de la fila 
    i+=1                        #paso a la siguiente fila
    
suma_cols = [0]*cols            #creamos un vector para guardar la suma de cada columna 

j=0                     #indice de columna

while j<cols:          #recorremos las filas una por una 
    suma = 0                #acumulador para la suma de la columna j 
    i=0                     #indice de filas 
    while i< filas:
        suma += matriz[i][j]
        i+=1
    suma_cols[j] = suma
    j+=1
    
valor_buscado = 7 #definimos el valor que quiero buscar
encontado = 0 #Bandera: 0 significa "no encontrado ", 1 significa "encontrado "

pos_fil =-1  #gaurda la fila donde aparece el valor ; -1 indice "no existe"
pos_col =-1  #gaurda la columna donde aparece el valor ; -1 indice "no existe"

i=0                                 #indice de filas 
while i<filas and encontado== 0: #se sigue buscando mientras queden filas y aun no lo alla encontrado
#---------------------------------------------------------

    j=0                                     #inicio en la primera columna de la fila i
    while j<cols and encontado== 0:         #sigue buscando mientras queden columnas y aun no lo alla encontrado
        if matriz[i][j] == valor_buscado:   #compara el valor actual con el valor buscado 
            encontado = 1                   #marco ya se encontro para detener los while 
            pos_col = j                     #guardo la columna 
            pos_fil = i                     #guardo la fila
        j+=1                                #paso a la siguiente columna
    i+=1                                    #paso a la siguiente fila
    
print("matriz: ",matriz)
print("Suma por filas:", suma_filas)
print("Suma por columnas:", suma_cols)   
print("posicion del",valor_buscado,":",(pos_fil,pos_col))


# promedios por filas 
promedio_filas = [0]*filas  #creamos un nuevo vector para gauradar el promedio de cada fila
i=0
while i<filas:          #recorremos las filas una por una 
    suma = 0                #acumulador para la suma de la fila i
    j=0                         #indice de columnas
    while j<cols:               #recprre todas las columanas de la fila i
        suma += matriz[i][j]    #sumo el elemento de la posicion (i,j)
        j+=1                    #paso a la siguente columna 
    promedio_filas[i]=suma/cols         #guardo el promedio de la fila 
    i+=1                        #paso a la siguiente fila
    
print("promedio por filas:", promedio_filas)

#Declare cuantos datos son pares y cuantos son impares
par = 0                  #contador de pares
impares = 0              # contador de impares 
i=0                      #indice de filas 


while i < filas:       #Recorremos las filas una por una 
    j = 0              #Indice de columnas 
    suma = 0           #Acumulador para la suma de las fila i 
    
    while j< cols:
        if(matriz[i][j] %2)==0:  #revisamos si el valor actual es par 
            par += 1                # entonces par y aumento el contador
        else:
            impares += 1     # entonces impar y aumento el contador
        j += 1                #paso a la siguente columna 
    i += 1               #paso a la siguiente fila
    
    
print("Cantidad de impares: ", impares)
print ("cantidad de pares",par)
```
