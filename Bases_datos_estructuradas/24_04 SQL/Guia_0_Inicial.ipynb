Guía paso a paso para integrar Jupyter Notebook con MySQL
0. Antes de empezar

Ustedes ya tienen:

VS Code
Python
Jupyter Notebook

Ahora van a instalar y configurar:

MySQL Community Server
MySQL Workbench
Un entorno virtual para el proyecto
Los paquetes de Python necesarios para conectarse a MySQL desde el notebook

1. Qué deben instalar exactamente
1.1. Instalen primero MySQL Community Server

Este es el componente obligatorio. Es el motor de base de datos; sin él, ustedes no tienen dónde crear bases de datos, tablas ni guardar información. La línea 8.4 LTS está publicada por MySQL como una versión estable de largo soporte, así que para una ruta formativa es una opción muy adecuada.

1.2. Instalen después MySQL Workbench

MySQL Workbench no es estrictamente obligatorio, pero sí es muy recomendable para el curso. Les permite crear bases de datos, abrir conexiones, ejecutar consultas SQL y ver tablas de forma visual. Para ustedes, sobre todo al inicio, facilita mucho entender qué está pasando antes de pasar al notebook. MySQL documenta Workbench como herramienta disponible para Windows y explica que puede instalarse mediante el instalador de Windows o con MSI independiente.

1.3. No necesitan instalar desde MySQL otros productos

Para esta etapa no necesitan MySQL Shell, Router, NDB, ODBC ni otros conectores desde la web de MySQL. Tampoco necesitan instalar Connector/Python desde la página de MySQL, porque ese conector lo van a instalar dentro de su entorno de Python con pip. MySQL lista esos productos por separado, pero para esta integración básica no hacen falta.


2. Paso 1: instalar MySQL Community Server

Busquen e instalen MySQL Community Server 8.4 LTS para Windows. MySQL documenta que, para Windows, la instalación puede hacerse con el MSI, y que el directorio por defecto de MySQL Server 8.4 usando MSI queda bajo C:\Program Files\MySQL\MySQL Server 8.4, aunque puede cambiarse.

Durante la instalación, presten atención a esto:
2.1. Tipo de instalación

Si el asistente les pregunta opciones, elijan la que instale el servidor de forma estándar. No hace falta personalizar demasiado en esta etapa.

2.2. Puerto

Dejen el puerto por defecto:  3306

Ese es el puerto habitual con el que luego se conectarán desde Workbench y desde Python. MySQL usa ese puerto por defecto para las conexiones clásicas del servidor.

3.3. Usuario administrador

Normalmente el usuario principal será: root

2.4. Contraseña

Definan una contraseña y anótenla muy bien. Esa misma contraseña la necesitarán después en Workbench y en Python.

2.5. Servicio de Windows

Si el instalador les pregunta si desean que MySQL arranque como servicio automáticamente, déjenlo activado. Eso evita que luego el servidor no esté disponible cuando abran Workbench o el notebook.

Qué debe pasar al final

La instalación debe dejarles el servidor listo y ejecutándose. Si MySQL Server no quedó corriendo, no podrán conectarse luego ni desde Workbench ni desde Python. La documentación de instalación en Windows describe precisamente el uso del MSI y la disposición del servidor en ese sistema.

3. Paso 2: instalar MySQL Workbench

Ahora instalen MySQL Workbench para Windows. MySQL Workbench puede instalarse con MySQL Installer o con el instalador MSI independiente para Windows. En la página oficial de descarga, MySQL también indica que en Windows hay prerrequisitos del sistema para que Workbench funcione correctamente.

Qué van a hacer con Workbench

Workbench les va a servir para:

abrir una conexión local al servidor,
crear la base de datos,
crear tablas,
ejecutar consultas,
y verificar si los datos sí quedaron guardados.
Qué deben revisar al abrirlo

Cuando Workbench abra, creen o usen una conexión local con estos datos:

Hostname: localhost
Port: 3306
Username: root

Luego escriben la contraseña que definieron al instalar MySQL Server. Workbench usa justamente ese esquema de conexión sobre host, puerto y usuario para conectarse al servidor local.
