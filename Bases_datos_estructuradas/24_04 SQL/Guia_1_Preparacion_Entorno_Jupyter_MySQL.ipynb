# Guía 1. Preparación del entorno para integrar Jupyter Notebook con MySQL

En esta guía ustedes van a dejar listo el entorno de trabajo para comenzar la integración entre **Jupyter Notebook** y **MySQL**.

Se parte de que ya tienen instalado:

- VS Code
- Python
- Jupyter Notebook

Lo que haremos aquí será:

1. instalar **MySQL Community Server**,
2. instalar **MySQL Workbench**,
3. crear la carpeta del proyecto,
4. crear el entorno virtual,
5. instalar los paquetes necesarios,
6. registrar el kernel del notebook,
7. verificar que todo quedó funcionando.


## 1. Qué deben instalar exactamente

Para esta ruta de trabajo ustedes necesitan:

### Obligatorio
- **MySQL Community Server**: es el motor de base de datos.

### Muy recomendable
- **MySQL Workbench**: les permite ver la base, crear tablas y ejecutar consultas de forma visual.

### No necesitan instalar desde MySQL
- MySQL Shell
- MySQL Router
- ODBC
- Otros conectores adicionales

El conector de Python lo vamos a instalar dentro del entorno virtual con `pip`.


## 2. Orden correcto de trabajo

Sigan este orden para evitar errores:

1. instalar MySQL Community Server,
2. instalar MySQL Workbench,
3. crear la carpeta del proyecto en VS Code,
4. crear el entorno virtual,
5. instalar paquetes,
6. registrar el kernel,
7. seleccionar el kernel en el notebook,
8. verificar imports y ruta del ejecutable.


## 3. Instalar MySQL Community Server

Durante la instalación de MySQL Server, revisen con cuidado estos puntos:

- dejen el puerto en `3306`,
- usen como usuario principal `root`,
- definan una contraseña y anótenla muy bien,
- dejen activado el inicio automático del servicio si el instalador lo ofrece.

Cuando terminen, MySQL debe quedar funcionando como servicio local.


## 4. Instalar MySQL Workbench

Luego instalen **MySQL Workbench**.

Lo van a usar para:

- abrir la conexión local,
- crear la base de datos,
- crear tablas,
- ejecutar consultas SQL,
- verificar si los datos sí se guardaron.

Al abrirlo, la conexión local normalmente usará:

- **Hostname**: `localhost`
- **Port**: `3306`
- **Username**: `root`


## 5. Crear la carpeta del proyecto en VS Code

Creen una carpeta exclusiva para este trabajo. Por ejemplo:

```text
integracion_jupyter_sql
```

Luego entren a esa carpeta desde VS Code con **File > Open Folder**.


## 6. Verificar qué Python usarán como base

Antes de crear el entorno virtual, ustedes deben saber cuál es el ejecutable de Python que van a usar.

En VS Code pueden hacer esto:

1. presionen `Ctrl + Shift + P`,
2. escriban **Python: Select Interpreter**, 
3. revisen la ruta del intérprete.

En este ejemplo voy a usar una ruta como esta:

```text
C:\Users\Hp\AppData\Local\Programs\Python\Python312\python.exe
```

En sus equipos puede cambiar el nombre del usuario o la versión.


## 7. Crear el entorno virtual

En la terminal de VS Code, dentro de la carpeta del proyecto, ejecuten:

```powershell
& "C:\Users\Hp\AppData\Local\Programs\Python\Python312\python.exe" -m venv .venv
```

### Qué hace este comando

- usa el Python base que ya encontraron,
- ejecuta el módulo `venv`,
- crea un entorno virtual llamado `.venv` dentro de la carpeta del proyecto.

Al terminar debe aparecer una carpeta llamada `.venv`.


## 8. Activar el entorno virtual

En PowerShell ejecuten:

```powershell
.\.venv\Scripts\Activate.ps1
```

Si la activación funciona, la terminal mostrará algo parecido a esto:

```text
(.venv)
```

### Si PowerShell bloquea la activación

Primero ejecuten:

```powershell
Set-ExecutionPolicy -Scope Process -ExecutionPolicy Bypass
```

y después repitan:

```powershell
.\.venv\Scripts\Activate.ps1
```


## 9. Actualizar `pip`

Con el entorno activado, ejecuten:

```powershell
.\.venv\Scripts\python.exe -m pip install --upgrade pip
```

Esto garantiza que el instalador de paquetes quede actualizado.


## 10. Instalar los paquetes necesarios

Ahora instalen los paquetes que sí van a usar en la integración:

```powershell
.\.venv\Scripts\python.exe -m pip install ipykernel mysql-connector-python sqlalchemy pandas
```

### Para qué sirve cada paquete

- `ipykernel`: permite que Jupyter use ese entorno como kernel.
- `mysql-connector-python`: abre la conexión entre Python y MySQL.
- `sqlalchemy`: les ayuda a crecer luego en conexiones más limpias.
- `pandas`: sirve para mostrar consultas SQL como tablas tipo `DataFrame`.


## 11. Registrar el entorno como kernel del notebook

Después de instalar los paquetes, registren el entorno como kernel:

```powershell
.\.venv\Scripts\python.exe -m ipykernel install --user --name py_sql --display-name "Python (py_sql)"
```

Esto hará que el notebook pueda mostrar ese entorno como opción en **Select Kernel**.


## 12. Seleccionar el kernel correcto en el notebook

Ahora abran su archivo `.ipynb` en VS Code.

Arriba a la derecha verán algo como **Select Kernel**.

Allí deben elegir:

- **Python (py_sql)**
- o el entorno Python cuya ruta apunte a `.venv`

### No deben elegir

- `MSSQL`
- servidores de Jupyter existentes

Lo correcto es trabajar con el entorno Python local que ustedes acaban de crear.


## 13. Verificar que el notebook sí quedó usando `.venv`

Ejecuten la siguiente celda. Lo importante aquí es que la ruta final apunte a `.venv`.


````python
import sys
print(sys.executable)
````

Si la salida termina en algo parecido a esto, todo va bien:

```text
...\integracion_jupyter_sql\.venv\Scripts\python.exe
```


## 14. Verificar que las librerías quedaron instaladas

Ahora ejecuten esta celda. Si no aparece error, el entorno ya quedó listo.


````python
import mysql.connector
import pandas as pd
from sqlalchemy import create_engine

print("Todo quedó instalado correctamente")
````

## 15. Verificación final en MySQL Workbench

Antes de pasar a la conexión con Python, entren a MySQL Workbench y prueben la conexión local con:

- host: `localhost`
- puerto: `3306`
- usuario: `root`
- contraseña: la que definieron al instalar MySQL

Si Workbench entra correctamente, eso significa que el servidor está funcionando y que ya pueden pasar a la primera integración real.
