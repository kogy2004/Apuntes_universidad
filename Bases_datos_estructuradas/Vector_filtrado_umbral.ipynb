Declare un vector de datos de prueba 

Defina un dato umbral de comparacion 

Declare un valor de filtradoque almacene datos (>umbral )

cuenta cuantos datos son impares 

```python
array = [4,7,10,3,8,9,2]
umbral = 6                          # defino el umbral 

n=  0                               # n representa el tamaño del vector 
for _ in array:                     #recorro el array para contar sus elementos
    n+=1                            # incrementa n en 1 por cada elemento del array

aux = [0]*n                         # creamos un vector auxiliar de tamaño n, para almacenar los valores filtrados 

k =0                                 # K representa la cantidad de elementos que cumplen la condicion

impares = 0                         # contador de numeros impares segun el vector original

i = 0                               # indice i para recorrer el array original con while (sin  range)

while i < n:                        # recorro el array original
    if(array[i] %2 )==0:            # revisamos si el valor actual es par 
        array[i]= array[i]*2            #si es par lo duplico y modifico el vector in-place
    else:
        impares +=1                 # entonces es impar y aumenta el contador 
        
    if array[i]>umbral:
        aux[k]= array[i]           
        k+=1                        # avanza en el siguiente indice
    i+=1                            

filtrado= [0]*k

i=0  #Reiniciamos i
while i<k:          #copiamos solo los k elementos validos desde aux hacia filtrados 
    filtrado[i] = aux[i]             #asignar el valor i-esimo
    i+=1
     
print("vector modificado: ", array)
print ("cantidad de impares" , impares )
print("vector filtrado (> umbral)", filtrado)
     
```
