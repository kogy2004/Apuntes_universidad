```python
# vector de prueba 
array = [1,2,3,4,5]

#Desplazamiento solicitado 
k=2
n=len(array)

#si el vector esta vacio, no hay nada que desplazar

if n== 0:
    print ("vector vacio: ", array)
else:
    #normalizamos k para evitar deplazamientos innecesarios
    #por ejemplo, k = 7 e n un vector de 5 equivale a k=7 (7 % 5)
    k=k%n
    
    #hacemos k veces un desplazamiento de i a la derecha
    for _ in range(k):
        #Guardamos el ultimo dato
        dato_ultimo = array[n-1]
        
        #corremos todos los datos una posicion a la derecha desde el final hacia el inicio 
        i=n-1
        while i>0:
            array[i]=array[i-1]
            i-=1
            
        # ponemos el ultimo en el primero 
        array[0]= dato_ultimo
    print("vector dsplazado: ",array)
```
