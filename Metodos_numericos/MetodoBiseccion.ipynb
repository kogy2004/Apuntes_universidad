```python
def funcionEvaluar (x):
  fx = 6*x**2-7*x+2
  return fx
```

```python
a = float(input("incerta a: "))
b = float(input("incerta b: "))
r= (a+b)/2

print("el valor de r es: ",r)
f_a= funcionEvaluar(a)
print("el valor de f(a) es: ",f_a)
f_b= funcionEvaluar(b)
print("el valor de f(b) es: ",  f_b)
f_a_x_F_b= f_a*f_b
print("el valor de f(a)*f(b) es: ",f_a_x_F_b)
f_r= funcionEvaluar(r)
print("el valor de f(r) es: ",f_r)
errorTeorico=(b-a)/2
print("el error teorico es: ",errorTeorico)

error = abs(f_r)
print("el error de f_r es:" , error)
```
